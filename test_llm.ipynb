```python
# 导入必要的库
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoConfig
import torch
from train import LLM, Config
```

```python
t = AutoTokenizer.from_pretrained('./saves/model')
# 注册自定义模型和配置
AutoConfig.register("small_model", Config)
AutoModelForCausalLM.register(Config, LLM)
```

```python
# 加载分词器
tokenizer = AutoTokenizer.from_pretrained("./tokenizer", use_fast=True)
```

```python
# 加载预训练模型
model = AutoModelForCausalLM.from_pretrained('./saves/model')
```

```python
input_data = [t.bos_token_id] + t.encode('你是人工智能吗')
print(input_data)
```

```python
for token in model.generate({"input_ids":torch.tensor(input_data).unsqueeze(0), "labels":None}, t.eos_token_id, 50, stream=False, temperature=0.0, top_k=8):
    print(t.decode(token[0]))
```
